### ***Feature Scaling***

```python
import numpy as np
```

```python
Test_Data=pd.DataFrame({'x':np.random.randint(-1,100,1000),
                        'y':np.random.randint(-50,50,1000) ,
                        'z':np.random.randint(-50,50,1000)})
```

```python
Test_Data
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
pl=plt.axes(projection='3d')
pl.scatter3D(Test_Data.x,Test_Data.y,Test_Data.z)
plt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
nm=Normalizer()
```

```python
Normal_Data=nm.fit_transform(Test_Data)
```

```python
Normal_Data
```

```python
Normal_df=pd.DataFrame(Normal_Data,columns=['A','B','C'])
```

```python
Normal_df
```

```python
pl=plt.axes(projection='3d')
pl.scatter3D(Normal_df.A,Normal_df.B,Normal_df.C)
plt.figure()
```

```python
data={"A":[1,2,4,5,6],
      "B":[10,40,np.nan,80,60],
      "C":[np.nan,np.nan,100,50,40]}
```

```python
DF=pd.DataFrame(data)
```

```python
DF.isnull()
```

```python
from sklearn.impute import SimpleImputer
```

```python
SM=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
SM.fit(DF)
```

```python
SM.transform(DF)
```
